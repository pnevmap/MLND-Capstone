```python
def find_Nan(value):
    return value!= value

header = ['Open', 'High', 'Low', 'Adj Close', 'Volume']
    
for i in header:
    print 'NaNs in ', i, ': ', df[i].apply(find_Nan).sum()
```

```python
df.plot(x='Date', y='Adj Close', title='M&S Adjusted Close Prices')
```

```python
df.plot(x='Date', y='Volume', title='M&S Trading Volumes')
```

```python
#feature engineering
df.loc[:,'Daily Var'] = df.loc[:,'High'] - df.loc[:,'Low']
df.loc[:, 'Daily Change'] = df.loc[:,'Close'] - df.loc[:,'Open']
df.loc[:, 'Daily Change %'] = df.loc[:, 'Daily Change'] / df.loc[:, 'Open'] * 100
```

```python
df.plot(x='Date', y='Daily Var', title='M&S Daily Variation of Price')
```

```python
df.plot(x='Date', y='Daily Change', title='M&S Daily Change of Price')
```

```python
df.plot(x='Date', y='Daily Change %', title='M&S Daily Change of Price Percents')
```
